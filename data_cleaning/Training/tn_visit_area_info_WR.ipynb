```python
file_path = r"E:\AI_HUB\data\Training\labeled_data\TL_csv\cleaned_data\tn_visit_area_info.csv"

tn_visit_area_info = pd.read_csv(file_path)

tn_visit_area_info
```

```python
# 데이터형 변환

tn_visit_area_info["VISIT_AREA_ID"] = tn_visit_area_info["VISIT_AREA_ID"].astype("object")

tn_visit_area_info["VISIT_START_YMD"] = pd.to_datetime(tn_visit_area_info["VISIT_START_YMD"])
tn_visit_area_info["VISIT_END_YMD"] = pd.to_datetime(tn_visit_area_info["VISIT_END_YMD"])

tn_visit_area_info["VISIT_MONTH"] = tn_visit_area_info["VISIT_START_YMD"].dt.month

int_columns = ["DGSTFN", "REVISIT_INTENTION", "RCMDTN_INTENTION"]
for col in int_columns:
    tn_visit_area_info[col] = tn_visit_area_info[col].astype("int")


# 변환된 데이터 확인
print(tn_visit_area_info.dtypes)
```

```python
type_cd = tn_visit_area_info['VISIT_AREA_TYPE_CD'].value_counts()
type_cd
```

```python
# VIS = 9, 21, 22, 23, 24인 데이터 삭제

tn_visit_area_info = tn_visit_area_info[~tn_visit_area_info["VISIT_AREA_TYPE_CD"].isin([9, 21, 22, 23, 24])]
```

```python
# VISIT_MONTH를 계절로 매핑
season_mapping = {
    12: "Winter", 1: "Winter", 2: "Winter",
    3: "Spring", 4: "Spring", 5: "Spring",
    6: "Summer", 7: "Summer", 8: "Summer",
    9: "Fall", 10: "Fall", 11: "Fall"
}

tn_visit_area_info["VISIT_SEASON"] = tn_visit_area_info["VISIT_MONTH"].map(season_mapping)
```

```python
# 필요 없는 열 제거
tn_visit_area_info_cleaned = tn_visit_area_info.drop(columns=["VISIT_START_YMD", "VISIT_END_YMD", "RESIDENCE_TIME_MIN", "VISIT_AREA_TYPE_CD", "REVISIT_YN", "VISIT_CHC_REASON_CD", 'VISIT_MONTH'], axis=1)
```

## WR(가중 평점) 구하기

```python
# v: 해당 관광지의 평가 개수 구하기
v = tn_visit_area_info_cleaned.groupby("VISIT_AREA_NM")["VISIT_AREA_ID"].count()

v
```

```python
# m: 최소 평가 개수(상위 90% 관광지의 평가 개수)
m = v.quantile(0.90)
```

```python
# W(가중 평점) 계산
w1, w2, w3 = 0.4, 0.3, 0.3
tn_visit_area_info_cleaned["W"] = (
    tn_visit_area_info_cleaned["DGSTFN"] * w1 +
    tn_visit_area_info_cleaned["REVISIT_INTENTION"] * w2 +
    tn_visit_area_info_cleaned["RCMDTN_INTENTION"] * w3
)
```

```python
# R: 관광지별 평균 만족도 계산
R = tn_visit_area_info_cleaned.groupby("VISIT_AREA_NM")["W"].mean()
```

```python
# C: 전체 평균 만족도 계산
C = tn_visit_area_info_cleaned["W"].mean()
```

```python
# WR(가중 평점) 계산
WR = ((v / (v + m)) * R) + ((m / (v + m)) * C)
```

```python
# WR을 기준으로 관광지 정렬 후 출력
WR = WR.sort_values(ascending=False)
WR.head(20)  # 상위 10개 관광지 출력
```

```python
# WR(가중 평점) 계산
WR = ((v / (v + m)) * R) + ((m / (v + m)) * C)

# WR을 원본 데이터에 병합
tn_visit_area_info_cleaned = tn_visit_area_info_cleaned.merge(WR.rename("WR"), on="VISIT_AREA_NM", how="left")

# 결과 확인
tn_visit_area_info_cleaned.head()
```
