```python
# 클러스터링에 필요한 정보만 추출

tn_traveller_master = tn_traveller_master[['TRAVELER_ID', 'GENDER', 'AGE_GRP', 'TRAVEL_STYL_1', 'TRAVEL_STYL_2', 'TRAVEL_STYL_3', 'TRAVEL_STYL_4', 'TRAVEL_STYL_5', 'TRAVEL_STYL_6','TRAVEL_STYL_7', 'TRAVEL_STYL_8', 'TRAVEL_MOTIVE_1', 'TRAVEL_MOTIVE_2', 'TRAVEL_MOTIVE_3', 'TRAVEL_COMPANIONS_NUM', 'TRAVEL_STATUS_ACCOMPANY', 'TRAVEL_STATUS_DESTINATION', 'TRAVEL_STATUS_YMD']]
```

```python
# 목적지가 제주인 데이터만 추출
tn_traveller_master_jeju_filtered = tn_traveller_master[tn_traveller_master['TRAVEL_STATUS_DESTINATION']=='제주']

tn_traveller_master_jeju_filtered
```

```python
# 이름 간결하게 변경
tn_traveller_jeju = tn_traveller_master_jeju_filtered
```

```python
tn_traveller_jeju = tn_traveller_jeju.drop(columns=['TRAVEL_STATUS_DESTINATION'])
```

```python
tn_traveller_jeju.isnull().sum()
```

```python
# 여행 동기 결측값이 전체 데이터의 5% 정도이므로 삭제함.
tn_traveller_jeju = tn_traveller_jeju.dropna(subset=['TRAVEL_MOTIVE_2', 'TRAVEL_MOTIVE_3'])
```

```python
tn_traveller_jeju.isnull().sum()
```

```python
# 데이터 타입 변환

tn_traveller_jeju['TRAVEL_MOTIVE_2'] = tn_traveller_jeju['TRAVEL_MOTIVE_2'].astype(int)
tn_traveller_jeju['TRAVEL_MOTIVE_3'] = tn_traveller_jeju['TRAVEL_MOTIVE_3'].astype(int)
```

```python
# TRAVEL_STATUS_TMD 

# 출발일(Start Date)과 종료일(End Date) 분리
tn_traveller_jeju[['Start_Date', 'End_Date']] = tn_traveller_jeju['TRAVEL_STATUS_YMD'].str.split('~', expand=True)

# 문자열 변환 후 출발일(Start Date)과 종료일(End Date) 분리
tn_traveller_jeju['TRAVEL_STATUS_YMD'] = tn_traveller_jeju['TRAVEL_STATUS_YMD'].astype(str)
tn_traveller_jeju[['Start_Date', 'End_Date']] = tn_traveller_jeju['TRAVEL_STATUS_YMD'].str.split('~', expand=True)

# 데이터 타입 변환
tn_traveller_jeju['Start_Date'] = pd.to_datetime(tn_traveller_jeju['Start_Date'], errors='coerce')
tn_traveller_jeju['End_Date'] = pd.to_datetime(tn_traveller_jeju['End_Date'], errors='coerce')

# 여행 기간(Duration) 계산
tn_traveller_jeju['Duration'] = (tn_traveller_jeju['End_Date'] - tn_traveller_jeju['Start_Date']).dt.days + 1
```

```python
# 'TRAVEL_STATUS_YMD' 열 삭제
tn_traveller_jeju.drop(columns=['TRAVEL_STATUS_YMD'], inplace=True)
```

```python
# 유지할 값 리스트 정의
valid_categories = ["자녀 동반 여행", "나홀로 여행", "2인 여행(가족 외)", "2인 가족 여행", "부모 동반 여행"]

# 조건을 기반으로 "기타" 처리
tn_traveller_jeju["ACCOMPANY_CATEGORY"] = tn_traveller_jeju["TRAVEL_STATUS_ACCOMPANY"].apply(
    lambda x: x if x in valid_categories else "기타"
)

# One-Hot Encoding 수행 (0과 1로 변환)
tn_traveller_jeju = pd.get_dummies(tn_traveller_jeju, columns=["ACCOMPANY_CATEGORY"])
```
