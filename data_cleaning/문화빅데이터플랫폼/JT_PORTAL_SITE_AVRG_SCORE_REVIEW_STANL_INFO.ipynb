```python
import pandas as pd
```

```python
file1_path = r"E:\문화빅데이터플랫폼\JT_PORTAL_SITE_AVRG_SCORE_REVIEW_STANL_INFO\JT_PORTAL_SITE_AVRG_SCORE_REVIEW_STANL_INFO_202306.csv"

file2_path = r"E:\문화빅데이터플랫폼\JT_PORTAL_SITE_AVRG_SCORE_REVIEW_STANL_INFO\JT_PORTAL_SITE_AVRG_SCORE_REVIEW_STANL_INFO_202312.csv"

df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)

# ✅ 행 기준으로 단순 합치기 (컬럼이 동일해야 함)
merged_df = pd.concat([df1, df2], ignore_index=True)
```
