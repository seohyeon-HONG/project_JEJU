```python
import pandas as pd
```

```python
file_path = r"E:\AI_HUB\data\Validation\labeled_data\VL_csv\tn_tour_photo_관광사진_H.csv"

tn_tour_photo = pd.read_csv(file_path)

tn_tour_photo
```

```python
tn_tour_photo.info()
```

```python
# dtype 변환

tn_tour_photo['VISIT_AREA_ID'] = tn_tour_photo['VISIT_AREA_ID'].astype(str)
tn_tour_photo['PHOTO_FILE_DT'] = pd.to_datetime(tn_tour_photo['PHOTO_FILE_DT'])
tn_tour_photo.info()
```

```python
# 필요 없는 열 제거

columns_to_drop = ['PHOTO_FILE_ID', 'PHOTO_FILE_FRMAT', 'PHOTO_FILE_SAVE_PATH', 'VISIT_AREA_NM', "PHOTO_FILE_RESOLUTION", "TOUR_PHOTO_SEQ"]

tn_tour_photo = tn_tour_photo.drop(columns=columns_to_drop)
tn_tour_photo.info()
```

## 제주도만 필터링

```python
# VISIT_AREA_NM을 알기 위해 병합 필요 -> tn_visit_area_info 테이블 불러오기(이미 제주도만 있음, 병합 안 되는 행은 삭제)

file_path = r"E:\AI_HUB\data\Validation\labeled_data\VL_csv\cleaned_data\tn_visit_area_info.csv"

tn_visit_area_info = pd.read_csv(file_path)
tn_visit_area_info.info()
```

```python
# 데이터 타입 변환
tn_visit_area_info['VISIT_AREA_ID'] = tn_visit_area_info['VISIT_AREA_ID'].astype(str)
```

```python
print(tn_tour_photo['VISIT_AREA_ID'].apply(type).value_counts())
print(tn_visit_area_info['VISIT_AREA_ID'].apply(type).value_counts())
```

```python
merged_data = pd.merge(tn_tour_photo, tn_visit_area_info, on='VISIT_AREA_ID', how='inner')

merged_data
```

```python
unique_count = tn_tour_photo['PHOTO_FILE_NM'].nunique()
total_count = tn_tour_photo['PHOTO_FILE_NM'].count()
print(f"Unique PHOTO_FILE_NM count: {unique_count}")
print(f"Total PHOTO_FILE_NM count: {total_count}")

# PHOTO_FILE_NM은 유일키다. 
```

```python
# PHOTO_FILE_NM 컬럼의 고유값 개수
unique_photo_count = merged_data['PHOTO_FILE_NM'].nunique()
print(f"Unique PHOTO_FILE_NM count: {unique_photo_count}")
```

```python
# 고유한 PHOTO_FILE_NM만 추출, 중복된 행은 삭제

merged_data_unique = merged_data.drop_duplicates(subset=['PHOTO_FILE_NM'], keep='first')
```

```python
# 필요한 열만 남겨서 저장
merged_data_cleaned = merged_data_unique[['VISIT_AREA_ID','VISIT_AREA_NM','PHOTO_FILE_NM', 'PHOTO_FILE_DT','PHOTO_FILE_X_COORD', 'PHOTO_FILE_Y_COORD']]

merged_data_cleaned
```

```python
import pandas as pd
import os
```

```python
file_path = r'E:\AI_HUB\data\Validation\labeled_data\VL_csv\cleaned_data\tn_tour_photo_jeju_filtered.csv'

# 관광사진 테이블에서 제주도만 추출한 데이터 불러오기
tn_tour_photo_jeju_filtered = pd.read_csv(file_path)
tn_tour_photo_jeju_filtered.head()
```

```python
# PHOTO_FILE_NM 리스트 생성
jeju_photo_names = tn_tour_photo_jeju_filtered['PHOTO_FILE_NM'].tolist()

# 원천 폴더 경로 지정 및 파일 목록 가져오기
jeju_raw_data = r"E:\AI_HUB\data\Validation\raw_data\VS_photo"
all_files_in_folder = os.listdir(jeju_raw_data)

# 삭제할 파일 리스트 생성 (경로를 문자열로 처리)
files_to_delete = [os.path.join(jeju_raw_data, file) for file in all_files_in_folder if file not in jeju_photo_names]

# 파일 삭제
for file_path in files_to_delete:
    try:
        os.remove(file_path)
        print(f"Deleted: {file_path}")
    except Exception as e:
        print(f"Error deleting {file_path}: {e}")

# 결과 확인
print(f"삭제된 파일 개수: {len(files_to_delete)}")
```
