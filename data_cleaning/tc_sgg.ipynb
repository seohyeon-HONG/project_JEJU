```python
import pandas as pd
```

```python
file_path = r"E:\AI_HUB\data\Training\labeled_data\TL_csv\tc_sgg_시군구코드.csv"

tc_sgg = pd.read_csv(file_path)

tc_sgg
```

```python
tc_sgg.info()
tc_sgg.isnull().sum()
```

```python
# SGG_CD2와 SGG_NM 결측 처리 -> 해당 행 삭제

sgg_data_cleaned = tc_sgg.dropna(subset=['SGG_CD2', 'SGG_NM'])
```

```python
# 시도, 시군구 필터링
sgg_data_cleaned = sgg_data_cleaned.drop(columns=['SGG_CD3', 'SGG_CD4', 'DONG_NM','RI_NM', 'SGG_CD'])

sgg_data_cleaned.info()
```

```python
sgg_data_cleaned
```

```python
unique_data = sgg_data_cleaned.drop_duplicates()
unique_data
```

```python
import shutil

# 파일 저장 경로
local_path = r"E:\AI_HUB\data\Training\labeled_data\TL_csv\cleaned_data\tc_sgg_filtered.csv"
unique_data.to_csv(local_path, index=False, encoding='utf-8-sig')

# 파일 저장 확인
print(f"로컬 경로에 파일 저장 완료: {local_path}")
```

```python
# 제주도 관련 데이터 필터링

jeju_filtered_data = tc_sgg[tc_sgg['SGG_NM'].str.contains("제주|서귀포", na=False)]

jeju_filtered_data.info()
jeju_filtered_data.head()
```

```python
jeju_filtered_data= jeju_filtered_data.dropna(subset=['SGG_CD3', 'DONG_NM'])
jeju_filtered_data_c = jeju_filtered_data.drop(columns=['RI_NM', 'SGG_CD4'])

jeju_filtered_data_c.info()
jeju_filtered_data_c.head()
```

```python
import shutil

# 파일 저장 경로
local_path = r"E:\AI_HUB\data\Training\labeled_data\TL_csv\cleaned_data\tc_sgg_jeju_filtered.csv"
jeju_filtered_data_c.to_csv(local_path, index=False, encoding='utf-8-sig')

# 파일 저장 확인
print(f"로컬 경로에 파일 저장 완료: {local_path}")
```
