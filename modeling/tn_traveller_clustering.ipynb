```python
import pandas as pd
```

```python
file_path = r"E:\AI_HUB\data\Training\labeled_data\TL_csv\tn_traveller_master_여행객 Master_H.csv"

tn_traveller_master = pd.read_csv(file_path)

tn_traveller_master
```

```python
# 클러스터링에 필요한 정보만 추출

tn_traveller_master = tn_traveller_master[['TRAVELER_ID', 'GENDER', 'AGE_GRP', 'TRAVEL_STATUS_ACCOMPANY', 'TRAVEL_STATUS_DESTINATION', 'TRAVEL_STATUS_YMD']]
```

```python
# 목적지가 제주인 데이터만 추출
tn_traveller_jeju = tn_traveller_master[tn_traveller_master['TRAVEL_STATUS_DESTINATION']=='제주']

tn_traveller_jeju
```

```python
tn_traveller_jeju = tn_traveller_jeju.drop(columns=['TRAVEL_STATUS_DESTINATION'], axis=1)
```

### 전처리

```python
tn_traveller_jeju.info()
```

```python
# TRAVEL_STATUS_TMD 

# 출발일(Start Date)과 종료일(End Date) 분리
tn_traveller_jeju[['Start_Date', 'End_Date']] = tn_traveller_jeju['TRAVEL_STATUS_YMD'].str.split('~', expand=True)

# 문자열 변환 후 출발일(Start Date)과 종료일(End Date) 분리
tn_traveller_jeju['TRAVEL_STATUS_YMD'] = tn_traveller_jeju['TRAVEL_STATUS_YMD'].astype(str)
tn_traveller_jeju[['Start_Date', 'End_Date']] = tn_traveller_jeju['TRAVEL_STATUS_YMD'].str.split('~', expand=True)

# 데이터 타입 변환
tn_traveller_jeju['Start_Date'] = pd.to_datetime(tn_traveller_jeju['Start_Date'], errors='coerce')
tn_traveller_jeju['End_Date'] = pd.to_datetime(tn_traveller_jeju['End_Date'], errors='coerce')

# 여행 기간(Duration) 계산
tn_traveller_jeju['Duration'] = (tn_traveller_jeju['End_Date'] - tn_traveller_jeju['Start_Date']).dt.days + 1
```

```python
# 'TRAVEL_STATUS_YMD' 열 삭제
tn_traveller_jeju = tn_traveller_jeju.drop(['Start_Date', 'End_Date'], axis=1)
```

```python
tn_traveller_jeju
```

```python
# GENDER 레이블 인코딩

tn_traveller_jeju['GENDER'] = tn_traveller_jeju['GENDER'].map({'남': 0, '여': 1})
```

```python
# AGE_GRP 레이블 인코딩

age_map = {20: 0, 30: 1, 40: 2, 50: 3, 60: 4}
tn_traveller_jeju['AGE_GRP'] = tn_traveller_jeju['AGE_GRP'].map(age_map)
```

```python
# 유지할 값 리스트 정의
valid_categories = ["자녀 동반 여행", "나홀로 여행", "2인 여행(가족 외)", "2인 가족 여행", "부모 동반 여행"]

# 조건을 기반으로 "기타" 처리
tn_traveller_jeju["ACCOMPANY_CATEGORY"] = tn_traveller_jeju["TRAVEL_STATUS_ACCOMPANY"].apply(
    lambda x: x if x in valid_categories else "기타"
)
```

```python
tn_traveller_jeju
```

```python
acc_cat_val_cnt = tn_traveller_jeju['ACCOMPANY_CATEGORY'].value_counts()
acc_cat_val_cnt
```

```python
# 고유값 출력 (현재 데이터가 어떻게 저장되어 있는지 확인)
print(tn_traveller_jeju["ACCOMPANY_CATEGORY"].unique())
```

```python
# ACCOMPANY_CATEGORY 공백 제거 & 표준화
tn_traveller_jeju["ACCOMPANY_CATEGORY"] = tn_traveller_jeju["ACCOMPANY_CATEGORY"].str.strip()

# 표준화
tn_traveller_jeju["ACCOMPANY_CATEGORY"] = tn_traveller_jeju["ACCOMPANY_CATEGORY"].replace({
    "2인 가족 여행 ": "2인 가족 여행",  
    "2인 여행(가족 외)": "2인 여행 (가족 외)",  
    "자녀 동반 여행 ": "자녀 동반 여행",
    "나홀로 여행 ": "나홀로 여행",
    "부모 동반 여행 ": "부모 동반 여행",
    "기타 ": "기타"
})

# One-Hot Encoding 수행 (dtype=int 설정하여 0/1 변환)
tn_traveller_jeju = pd.get_dummies(tn_traveller_jeju, columns=["ACCOMPANY_CATEGORY"], dtype=int)

# 결과 확인
print(tn_traveller_jeju.head())
```

```python
tn_traveller_jeju
```

```python
tn_traveller_jeju = tn_traveller_jeju.drop(tn_traveller_jeju[['TRAVEL_STATUS_ACCOMPANY', 'TRAVEL_STATUS_YMD']], axis=1)
```

```python
tn_traveller_jeju
```

```python
tn_traveller_jeju = tn_traveller_jeju.drop(columns=["TRAVELER_ID"], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
tn_traveller_jeju["Duration"] = scaler.fit_transform(tn_traveller_jeju[["Duration"]])
```

```python
tn_traveller_jeju
```

```python
from sklearn.cluster import KMeans

# 일단 4로 해 봄
kmeans = KMeans(n_clusters=4, random_state=42)
tn_traveller_jeju["cluster"] = kmeans.fit_predict(tn_traveller_jeju)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# PCA를 사용하여 2차원으로 차원 축소
pca = PCA(n_components=2)
principal_components = pca.fit_transform(tn_traveller_jeju.drop(columns=["cluster"]))  # 클러스터 컬럼 제외

# PCA 결과를 데이터프레임으로 변환
tn_traveller_jeju["PCA1"] = principal_components[:, 0]
tn_traveller_jeju["PCA2"] = principal_components[:, 1]

# 군집 시각화
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tn_traveller_jeju, x="PCA1", y="PCA2", hue="cluster", palette="tab10", alpha=0.7)
plt.title("PCA-based Clustering Visualization")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Cluster")
plt.show()
```

```python
print(principal_components.shape)
```

```python
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt

# t-SNE 변환
tsne = TSNE(n_components=2, random_state=42)
tsne_components = tsne.fit_transform(tn_traveller_jeju.drop(columns=["cluster"]))

# 시각화
tn_traveller_jeju["tSNE1"] = tsne_components[:, 0]
tn_traveller_jeju["tSNE2"] = tsne_components[:, 1]

plt.figure(figsize=(8,6))
sns.scatterplot(data=tn_traveller_jeju, x="tSNE1", y="tSNE2", hue="cluster", palette="tab10", alpha=0.7)
plt.title("t-SNE-based Clustering Visualization")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend(title="Cluster")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(tn_traveller_jeju.drop(columns=["cluster"]), tn_traveller_jeju["cluster"])
print(f"Silhouette Score: {sil_score:.3f}")
```

K=4로는 잘 군집화가 안 됨. 

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

inertia_values = []
K_range = range(2, 10)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tn_traveller_jeju.drop(columns=["cluster"]))
    inertia_values.append(kmeans.inertia_)

# 그래프 시각화
plt.figure(figsize=(6,4))
plt.plot(K_range, inertia_values, marker="o", linestyle="-")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K")
plt.show()

# K=5로 c=1 주황색 군집을 세분화할 가능성. 
```

```python
from sklearn.metrics import silhouette_score

# K=5로 K-Means 클러스터링 적용
kmeans = KMeans(n_clusters=5, random_state=42)
tn_traveller_jeju["cluster"] = kmeans.fit_predict(tn_traveller_jeju.drop(columns=["cluster"], errors='ignore'))

# 실루엣 점수 확인
sil_score = silhouette_score(tn_traveller_jeju.drop(columns=["cluster"]), tn_traveller_jeju["cluster"])
print(f"Silhouette Score (K=5): {sil_score:.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 원핫 인코딩된 동반자 유형 컬럼 선택
accompany_cols = [col for col in tn_traveller_jeju.columns if "ACCOMPANY_CATEGORY_" in col]

# 클러스터별 동반자 유형 합계 계산
accompany_cluster_counts = tn_traveller_jeju.groupby("cluster")[accompany_cols].sum()

# 데이터 변환 (Seaborn에서 사용 가능하도록 Melt 변환)
accompany_cluster_melted = accompany_cluster_counts.reset_index().melt(id_vars="cluster", var_name="Accompany_Type", value_name="Count")

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x="cluster", y="Count", hue="Accompany_Type", data=accompany_cluster_melted)
plt.title("Accompany Category Distribution by Cluster")
plt.show()
```

* 0은 기타, 나홀로
* 1은 가족 외 2인 -> 친구, 연인
* 2는 2인 가족, 부모동반, 자녀동반
* 3은 가족 외 2인
* 4는 나홀로
* 5는 혼재됨

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x="cluster", y="AGE_GRP", data=tn_traveller_jeju, inner="quartile")
plt.title("클러스터별 연령 그룹 분포 (Violin Plot)")
plt.show()
```

* 0 - 젊은 층, 혼재
* 1 = 20대가 주
* 2 - 다양한 분포, 중년층 포함
* 3 - 청년+중장년
* 4 - 20대가 주
* 5 - 4~60대가 주

```python
plt.figure(figsize=(8, 5))

# KDE Plot을 클러스터별로 그리기
for cluster in tn_traveller_jeju["cluster"].unique():
    sns.kdeplot(tn_traveller_jeju[tn_traveller_jeju["cluster"] == cluster]["AGE_GRP"], label=f"Cluster {cluster}", shade=True)

plt.title("클러스터별 연령 그룹 분포 (KDE Plot)")
plt.xlabel("AGE_GRP")
plt.ylabel("Density")
plt.legend()
plt.show()
```
